```python
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/Dataset .csv")  # Replace with your actual filename
```

```python
# Count frequency of each price range
print(df['Price range'].value_counts())

# Visualize
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Price range', data=df)
plt.title("Most Common Price Range")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```

```python
avg_rating_per_price = df.groupby('Price range')['Aggregate rating'].mean().reset_index()
print(avg_rating_per_price)

# Bar plot
sns.barplot(x='Price range', y='Aggregate rating', data=avg_rating_per_price)
plt.title("Average Rating by Price Range")
plt.ylabel("Average Rating")
plt.show()
```

```python
avg_rating_color = df.groupby('Rating color')['Aggregate rating'].mean().sort_values(ascending=False)
print(avg_rating_color)

# Optional visualization
sns.barplot(x=avg_rating_color.values, y=avg_rating_color.index)
plt.title("Average Rating by Color")
plt.xlabel("Average Rating")
plt.ylabel("Rating Color")
plt.show()
```
