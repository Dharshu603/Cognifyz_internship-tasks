```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/Dataset .csv")  # Replace with your actual file path

# Clean column names
df.columns = df.columns.str.strip()

# Use the correct rating column
rating_col = 'Aggregate rating'

# Convert to numeric in case of any non-numeric values
df[rating_col] = pd.to_numeric(df[rating_col], errors='coerce')

# Drop missing values in rating column
df_clean = df.dropna(subset=[rating_col])

# --- Plot 1: Histogram of Ratings ---
plt.figure(figsize=(8, 5))
sns.histplot(df_clean[rating_col], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()

# --- Plot 2: Boxplot of Ratings ---
plt.figure(figsize=(6, 4))
sns.boxplot(x=df_clean[rating_col], color='lightgreen')
plt.title("Boxplot of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.show()

# --- Plot 3: Average Rating by Cuisine ---
if 'Cuisines' in df.columns:
    top_cuisines = df_clean.groupby('Cuisines')[rating_col].mean().sort_values(ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='magma')
    plt.title("Average Aggregate Rating by Cuisine (Top 10)")
    plt.xlabel("Average Rating")
    plt.ylabel("Cuisine")
    plt.show()

# --- Plot 4: Average Rating by City ---
if 'City' in df.columns:
    top_cities = df_clean.groupby('City')[rating_col].mean().sort_values(ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
    plt.title("Average Aggregate Rating by City (Top 10)")
    plt.xlabel("Average Rating")
    plt.ylabel("City")
    plt.show()

# --- Plot 5: Price Range vs Rating ---
if 'Price range' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Price range', y=rating_col, data=df_clean)
    plt.title("Price Range vs Aggregate Rating")
    plt.xlabel("Price Range")
    plt.ylabel("Aggregate Rating")
    plt.show()

# --- Plot 6: Online Delivery vs Rating ---
if 'Has Online delivery' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Has Online delivery', y=rating_col, data=df_clean)
    plt.title("Online Delivery vs Aggregate Rating")
    plt.xlabel("Has Online Delivery")
    plt.ylabel("Aggregate Rating")
    plt.show()

# --- Plot 7: Table Booking vs Rating ---
if 'Has Table booking' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Has Table booking', y=rating_col, data=df_clean)
    plt.title("Table Booking vs Aggregate Rating")
    plt.xlabel("Has Table Booking")
    plt.ylabel("Aggregate Rating")
    plt.show()
```
