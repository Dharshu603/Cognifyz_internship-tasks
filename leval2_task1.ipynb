```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (change filename as needed)
df = pd.read_csv("/content/Dataset .csv")  # Replace with your file name

# Clean column names: lowercase and underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# View cleaned column names
print("Cleaned column names:", df.columns.tolist())
```

```python
# Table Booking Percentage
table_booking_percent = df['has_table_booking'].value_counts(normalize=True) * 100
print("Table Booking (%):\n", table_booking_percent)

# Online Delivery Percentage
online_delivery_percent = df['has_online_delivery'].value_counts(normalize=True) * 100
print("\nOnline Delivery (%):\n", online_delivery_percent)

# Pie Chart for Table Booking
table_booking_percent.plot(kind='pie', autopct='%1.1f%%', title='Table Booking Availability', ylabel="")
plt.show()

# Pie Chart for Online Delivery
online_delivery_percent.plot(kind='pie', autopct='%1.1f%%', title='Online Delivery Availability', ylabel="")
plt.show()
```

```python
avg_rating_booking = df.groupby('has_table_booking')['aggregate_rating'].mean().reset_index()

# Display
print("\nAverage Ratings (Table Booking):\n", avg_rating_booking)

# Bar Plot
sns.barplot(x='has_table_booking', y='aggregate_rating', data=avg_rating_booking)
plt.title("Average Rating: Table Booking vs No Booking")
plt.xlabel("Has Table Booking")
plt.ylabel("Average Rating")
plt.show()
```

```python
delivery_by_price = df.groupby(['price_range', 'has_online_delivery']).size().unstack(fill_value=0)

# Display counts
print("\nOnline Delivery by Price Range:\n", delivery_by_price)

# Stacked Bar Chart
delivery_by_price.plot(kind='bar', stacked=True)
plt.title("Online Delivery by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.legend(title="Has Online Delivery")
plt.show()
```
